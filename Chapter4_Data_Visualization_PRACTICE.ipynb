## 1. Line Chart: Population Growth

```python
# Import libraries
import matplotlib.pyplot as plt

# Define variables
Year = [2010, 2012, 2014, 2016, 2018, 2020]
Population = [9.5, 10.2, 10.8, 11.4, 12.0, 12.6]  # Population in millions

# Plot the line chart
plt.figure(figsize=(8, 5))
plt.plot(Year, Population, marker='o', color='b', label='Population')
plt.title('Population Growth Over Time')
plt.xlabel('Year')
plt.ylabel('Population (millions)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

## ðŸ“ˆ Interpretation: Population Growth Over Time

The chart shows a steady increase in population from 9.5 million in 2010 to 12.6 million in 2020.

Growth is consistent and gradual, with an average increase of about 0.5â€“0.6 million every two years.

This upward trend suggests sustained demographic expansion, which may reflect higher birth rates, improved healthcare, or migration trends.

For stakeholders, this indicates rising demand for housing, education, healthcare, and infrastructure over time.

```python


# Sample GDP data by region (in trillions of dollars)
regions = ["North America", "Europe", "Asia", "Africa", "South America", "Oceania"]
gdp_values = [25, 22, 30, 3, 4, 2]

# Create the bar chart
plt.figure(figsize=(8, 5))
plt.bar(regions, gdp_values, color="skyblue", edgecolor="black")

# Add titles and labels
plt.title("GDP by Region (in Trillions $)", fontsize=14, fontweight="bold")
plt.xlabel("Region", fontsize=12)
plt.ylabel("GDP (Trillions $)", fontsize=12)

# Rotate X-axis labels for readability
plt.xticks(rotation=30, ha="right")

# Show the chart
plt.tight_layout()
plt.show()
```

### ðŸ“Š Interpretation: GDP by Region (in Trillions $)

Asia ($30T) has the largest GDP, followed closely by North America ($25T) and Europe ($22T) â€” together these regions account for the majority of global economic output.

South America ($4T) and Africa ($3T) contribute much less, showing a significant economic gap compared to the major regions.

Oceania ($2T) has the smallest share, reflecting its smaller population and market size.

Overall, the chart highlights global economic imbalance, where a few regions dominate GDP, while others contribute relatively little.

## 3. Histogram: Exam Scores

```python
# Import libraries
import numpy as np

# Define variable
Scores = np.random.randint(40, 101, 100)  # Random scores between 40 and 100

# Plot the histogram
plt.figure(figsize=(8, 5))
plt.hist(Scores, bins=10, color='teal', edgecolor='black')
plt.title("Distribution of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## ðŸ“Š Interpretation: Distribution of Exam Scores

The histogram groups 100 exam scores into ranges (bins of 10 points each).

The chart shows where most studentsâ€™ scores fall, making it easy to spot concentration areas (e.g., if many students are in the 70â€“80 range).

Taller bars = higher frequency of students scoring within that interval.

If the shape is bell-like, it suggests a normal distribution where most students score around the average.

If skewed, it highlights performance imbalance â€” for example, more students clustering toward higher or lower ends.

## 4. Scatter Plot: Reading Hours vs. Grade

```python
# Define variables
Reading_Hours = np.random.randint(1, 11, 100)  # Random hours between 1 and 10
Grade = np.random.randint(40, 101, 100)  # Random grades between 40 and 100

# Plot the scatter plot
plt.figure(figsize=(8, 5))
plt.scatter(Reading_Hours, Grade, color='orange', label='Data Points')
plt.title("Reading Hours vs. Grade")
plt.xlabel("Reading Hours")
plt.ylabel("Grade")
plt.legend()
plt.tight_layout()
plt.show()
```

## ðŸ“ˆ Interpretation: Reading Hours vs. Grade

Each dot represents a studentâ€™s study time (X-axis) and exam grade (Y-axis).

A positive trend is expected: as reading hours increase, grades generally tend to improve.

However, scatter plots also reveal variability â€” some students with fewer hours may still perform well, while others with more hours may not, highlighting that effort is important but not the only factor.

This visualization is useful for identifying correlations between study habits and academic performance, and for detecting outliers (e.g., low grades despite high study hours).

```python

import matplotlib.pyplot as plt
import numpy as np

# Scatter plot example
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)

plt.figure(figsize=(6,4))
plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, edgecolors="w")
plt.title("Scatter Plot Example")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

## ðŸ“ˆ Interpretation: Scatter Plot Example

Each point represents a data observation with:

X values on the horizontal axis

Y values on the vertical axis

Color and size adding extra dimensions of information

The plot shows how data points are distributed across two variables, useful for detecting relationships, clusters, or outliers.

Because the data here is random, no clear correlation appears â€” but in real cases, scatter plots help identify whether variables move together (positive correlation), opposite (negative correlation), or show no relationship.

This type of visualization is especially powerful when analyzing multidimensional data, since it encodes more than two features (position, color, size).

```python

# Bar chart example
categories = ["A", "B", "C", "D"]
values = [23, 17, 35, 29]

plt.figure(figsize=(6,4))
plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

## ðŸ“Š Interpretation: Bar Chart Example

Category C (35) has the highest value, indicating it contributes the most among the four groups.

Category B (17) is the lowest, showing the weakest performance or smallest share.

Category A (23) and Category D (29) fall in the middle, with D closer to the higher end.

Overall, the chart makes it easy to compare differences between categories at a glance, highlighting where resources, performance, or attention are most concentrated.

```python

# Pie chart example
sizes = [15, 30, 45, 10]
labels = ["Category A", "Category B", "Category C", "Category D"]
explode = (0, 0.1, 0, 0)  # explode 2nd slice

plt.figure(figsize=(6,6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

## ðŸ¥§ Interpretation: Pie Chart Example

Category C (45%) is the largest share, making it the dominant contributor.

Category B (30%) is the second largest, emphasized with an â€œexplodedâ€ slice to draw stakeholder attention.

Category A (15%) and Category D (10%) are much smaller, together accounting for only a quarter of the total.

Overall, the chart highlights the imbalanced distribution, where a single category (C) takes nearly half of the whole, while the others have smaller contributions.

```python

# Line chart example
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(6,4))
plt.plot(x, y, label="Sine Wave")
plt.title("Line Chart Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

## ðŸ“ˆ Interpretation: Line Chart Example (Sine Wave)

The chart visualizes a continuous mathematical function (sine) across values of x from 0 to 10.

The wave pattern shows cyclical fluctuations â€” rising, peaking, falling, and repeating.

In real-world data, similar line charts are used to show time series trends like stock prices, sales performance, or seasonal demand.

The smooth curve makes it easy to identify patterns, cycles, and direction of change over time or across variables.
